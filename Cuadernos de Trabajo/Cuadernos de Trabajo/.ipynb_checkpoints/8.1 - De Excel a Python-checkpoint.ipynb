## Operaciones en Pandas

Vamos creando una columan nueva de ganancias que será igual a precio menos costo de precios y costos

```python
precios_['Precio'] = precios_['Precio'].str.replace('$','').astype('float')
```

```python
precios_
```

```python
precios_['Costo'] = precios_['Costo'].str.replace('$','').astype('float')
```

```python
precios_
```

```python
precios_['Ganancias'] = precios_['Precio'] - precios_['Costo']
```

```python
precios_
```

Arma un nuevo dataframe con todos los platillos de más de 100 pesos de precio (usando filtros)

```python
nprecios = precios_[precios_['Precio'] > 100]
```

```python
nprecios.head(1)
```

Arma una nueva columna que clasifique platillos en alto/bajo margen, según si su ganancia es mayor a 100 pesos

```python
precios_['Margen'] = np.where(precios_['Ganancias']>100,'alto','bajo')
```

## Merging o Vlookup

Vamos a juntar los 2 dataframes usando un left join

```python
restprecio = datos_.merge(precios_,on=['Producto'],how='left')
```

## Tablas Dinámicas

Pandas también nos permite analizar datos mediante tablas dinámicas

Vamos creando una tabla dinámica que nos muestre las ventas por categoría

```python
venxcat = pd.pivot_table(restprecio, values='Precio',index='Categoria',aggfunc=np.sum)
```

```python
venxcat
```

Ahora una tabla dinámica que nos muestre las ventas por categoría y por tipo en la columna

```python
venxcatxtipo = pd.pivot_table(restprecio, values='Precio',index='Categoria', columns='Tipo_x', aggfunc=np.sum)
```

```python
venxcatxtipo
```

Crea una tabla dináica que muestre las ventas por tipo y categoría (filas), y por mesero (columnas)

```python
venxcatxtipo = pd.pivot_table(restprecio, values='Precio',index=['Tipo_x','Categoria'], columns='Mesero', aggfunc=np.sum)
```

```python
venxcatxtipo
```

Crea una tabla dinámica que muestre las ventas promedio por mesero

```python
venxmesero = pd.pivot_table(restprecio,values='Precio',index='Mesero') #promedio por default
```

```python
venxmesero
```
